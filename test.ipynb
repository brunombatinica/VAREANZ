```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


v1 = 5
v2 = 5

a1 = np.random.normal(0, np.sqrt(v1), 1000000)
a2 = np.random.normal(0, np.sqrt(v2), 1000000)

diff = a1 - a2
diff.std()
```

```python
# plot a1 and a2
plt.hist(a1, alpha=0.5, label='a1')
plt.hist(a2, alpha=0.5, label='a2')
plt.vlines(a1.mean(), a2.mean(), 500, color='black', linestyle='--')
plt.legend()
plt.show()

```
